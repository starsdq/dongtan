# 03. 경쟁사 POI 수집 및 검증

**목적**: `data/competitors.json` 의 좌표·정보를 검증하고 최신화  
**방법**: 카카오맵 로컬 API (REST) 또는 네이버 지도 검색으로 좌표 확인

> `data/competitors.json` 에는 이미 주요 시설 데이터가 있습니다.  
> 이 노트북은 좌표 정확도 검증 및 신규 시설 추가에 사용합니다.

````python
import requests
import json
from pathlib import Path

DATA_DIR = Path('../data')

# 카카오맵 REST API 키 (선택)
# https://developers.kakao.com → 앱 만들기 → REST API 키
KAKAO_REST_KEY = 'YOUR_KAKAO_REST_API_KEY'
````

## Step 1: 현재 competitors.json 확인

````python
with open(DATA_DIR / 'competitors.json', encoding='utf-8') as f:
    comp_data = json.load(f)

for f in comp_data['facilities']:
    print(f"{f['name']:30s} | {f['category']:12s} | {f['lat']}, {f['lng']}")
````

## Step 2: 카카오 키워드 검색으로 좌표 검증

````python
def search_kakao(query, x=127.077, y=37.207, radius=50000):
    """카카오맵 키워드 검색 API"""
    url = 'https://dapi.kakao.com/v2/local/search/keyword.json'
    headers = {'Authorization': f'KakaoAK {KAKAO_REST_KEY}'}
    params = {'query': query, 'x': x, 'y': y, 'radius': radius, 'size': 3}
    resp = requests.get(url, headers=headers, params=params, timeout=10)
    return resp.json()

# 예시: 롯데백화점 동탄점 좌표 검증
# result = search_kakao('롯데백화점 동탄점')
# for doc in result.get('documents', []):
#     print(f"{doc['place_name']}: {doc['y']}, {doc['x']}")

print('카카오 REST API 키 설정 후 주석 해제하여 실행')
````
